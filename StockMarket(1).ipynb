# Portafolio Nariz o de Minima Varianza

```python
n_acciones = len(acciones_a_invertir)
df = precios['Adj Close']
```

```python
def LogRendimientos(df):
    LogR = np.log(df).diff()
    LogR.reset_index(inplace=True)
    LogR.drop(0,inplace=True)
    LogR.fillna(method='ffill', inplace=True)
    return LogR

logR = LogRendimientos(df)

r = logR.mean()
V = logR.cov()
V_inv = np.linalg.inv(np.matrix(V))
```

```python
#Portafolio Nariz
e = np.matrix(np.ones(n_acciones))
w = V_inv.dot(e.transpose())/(e.dot(V_inv.dot(e.transpose())))
```
